```python
import requests

from matplotlib import pyplot as plt
import pandas as pd
import time
import numpy as np 
%matplotlib inline
```

```python
def get_location():
    r = requests.get(url='http://api.open-notify.org/iss-now.json')
    location = (r.json())
    
    x_list=[]
    
    for i in range(101):
        r = requests.get(url='http;//api.open-notify.org/iss-now.json')
        location = (r.json())
        latitude = float(location ['iss_position']['latitude'])
        x_list.append(latitude)
        time.sleep(10)

    y_list=[]
   
    for i in range(101):
        r = requests.get(url='http;//api.open-notify.org/iss-now.json')
        location = (r.json())
        longitude = float(location['iss_position']['longitude'])
        y_list.append(longitude)
        time.sleep(10)

    z_list=[]
    
    for i in range(101):
        r = requests.get(url='http;//api.open-notify.org/iss-now.json')
        location = (r.json())
        timestamp = float(location["timestamp"])
        z_list.append(timestamp)
        time.sleep(10)
        
    m_list=list(zip(x_list, y_list, z_list))
    print(m_list)
```

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure()
fig.set_size_inches(20, 19)

ax = plt.axes(projection ='3d')

zdata = [z_list]
xdata = [x_list]
ydata = [y_list]
ax.scatter3D(xdata, ydata, zdata, cmap='Blues')
```
